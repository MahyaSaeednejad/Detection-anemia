```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.fit_transform(y_test)
```

```python
model = Sequential()
K.set_image_data_format('channels_last')

model.add(Conv2D(64, (9, 9), padding= 'same', input_shape=(9, 96, 1), activation= 'relu' )) #9*96*30
model.add(MaxPooling2D(pool_size=(2, 2))) #4*48*20
model.add(Conv2D(32, (4, 4), activation= 'relu' )) #2*46*15
model.add(MaxPooling2D(pool_size=(1, 2))) #1*23*15
model.add(Conv2D(16, (1, 5), padding= 'same', activation= 'relu' ))
model.add(Dropout(0.1))
model.add(Flatten())
model.add(Dense(128, activation= 'relu' ))
model.add(Dense(50, activation= 'relu' ))
model.add(Dense(4, activation= 'linear' ))

opt =optimizers.Adadelta(lr =0.1)

  # Compile model
model.compile(loss= 'categorical_crossentropy' , optimizer= 'Adam' , metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=300, batch_size= 160)
score = model.evaluate(X_test, y_test, batch_size=160)
```
