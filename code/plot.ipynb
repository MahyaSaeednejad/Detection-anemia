```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("X_HW_dataset/sample_1.csv")

y = df['Unnamed: 0']
X = df.drop(columns=['Unnamed: 0'])
fig= plt.figure(figsize=(20,10))

for index, row in X.iterrows():
    plt.plot(X.columns[:],row,label=str(y[index]))
    
plt.ylabel('intensity')
plt.xlabel('wavelength')
plt.legend(y, bbox_to_anchor=(1, 1))
plt.show()
```
